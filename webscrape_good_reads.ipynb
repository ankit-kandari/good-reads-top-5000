```python
name = []
writer = []
rating = []

for i in range(50):
    url = "https://www.goodreads.com/list/show/19.Best_for_Book_Clubs?page={}".format(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.content,"lxml")
    boxes = soup.find_all('td', {'width': '100%', 'valign': 'top'})
    names = soup.find_all('span', {'itemprop': 'name'})
    for i in range(1,len(names),2):
      writer.append(names[i].text.strip())

    title = soup.find_all("span",{'itemprop': 'name', 'role': 'heading', 'aria-level': '4'})
    for t in title:
      name.append(t.text.strip())
    rate = soup.find_all("span",class_="minirating")
    for j in rate:
      rating.append(j.text.strip())
```

```python
good_read = pd.DataFrame({"book":name,
              "author":writer,
              "rating":rating})
```

```python
good_read.to_csv("goodreads_book.csv",index = False)
```
